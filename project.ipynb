```python
import pickle
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('loan_detection.csv')
df
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(keep='first',inplace=True)
```

```python
df
```

```python
df['Loan_Status_label'].value_counts()
```

```python
from sklearn.utils import resample
rejected = df[df['Loan_Status_label']==0]
approved = df[df['Loan_Status_label']==1]
rejected_down = resample(rejected,n_samples=len(approved),random_state=42,replace=False)
df = pd.concat([rejected_down,approved])
df
```

```python
df['Loan_Status_label'].value_counts()
```

```python
corr_target = df.corr()["Loan_Status_label"].sort_values(ascending=False)
print(corr_target)
```

```python
important_features = corr_target[abs(corr_target)>0.05]
important_features.drop('Loan_Status_label',inplace=True)
important_features.index
```

```python
X = df[important_features.index]
y = df['Loan_Status_label']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler

sr = StandardScaler()
```

```python
X_train = sr.fit_transform(X_train)
X_test = sr.transform(X_test)
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
print("Training Score: ",lr.score(X_train,y_train))
print("Testing Score",lr.score(X_test,y_test))
```

```python

dt = DecisionTreeClassifier(max_depth=9)
dt.fit(X_train,y_train)
print(i,"Training Score: ",dt.score(X_train,y_train))
print(i,"Testing Score",dt.score(X_test,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=9)
rf.fit(X_train,y_train)
```

```python
print("Training Score: ",rf.score(X_train,y_train))
print("Testing Score",rf.score(X_test,y_test))
```

```python
y_pred_train = rf.predict(X_train)
y_pred_test = rf.predict(X_test)
```

```python
print(confusion_matrix(y_train, y_pred_train))
```

```python
print(classification_report(y_train,y_pred_train))
```

```python
print(confusion_matrix(y_test, y_pred_test))
```

```python
print(classification_report(y_test, y_pred_test))
```

```python
from sklearn.metrics import roc_auc_score

probs = rf.predict_proba(X_test)[:,1]
print("AUC:",roc_auc_score(y_test, probs))
```
